```python
import numpy as np
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler, normalize 
from sklearn.metrics import silhouette_score 
import scipy.cluster.hierarchy as shc 
from sklearn.decomposition import PCA 
```

```python
data = pd.read_csv('customers.csv')
```

```python
data.head()
```

```python
data.drop('Region',inplace=True,axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data_scaled = StandardScaler().fit_transform(data)

data_normal = normalize(data_scaled)

data_normal = pd.DataFrame(data_normal)

data_normal
```

```python
# pca = PCA(n_components = 3) 
pca = PCA(n_components = 2) 
data_principal = pca.fit_transform(data_normal) 
data_principal = pd.DataFrame(data_principal) 
# data_principal.columns = ['P1', 'P2', 'P3'] 
data_principal.columns = ['P1', 'P2'] 
```

```python
plt.figure(figsize=(16,9))
plt.title('DendoGram')
Dendrogram = shc.dendrogram((shc.linkage(data_principal, method ='ward'))) 
```

```python
silhouette_scores = [] 
for clcomp in range(2,7):
    ac = AgglomerativeClustering(n_clusters = clcomp) 
    # Visualizing the clustering 
    plt.figure(figsize =(6, 6)) 
    plt.title('For {} Clusters'.format(clcomp))
#     plt.scatter(data_principal['P1'], data_principal['P2'], s = data_principal['P3'], c = ac.fit_predict(data_principal), cmap ='rainbow') 
    plt.scatter(data_principal['P1'], data_principal['P2'], c = ac.fit_predict(data_principal), cmap ='rainbow') 
    plt.show()
    silhouette_scores.append(silhouette_score(data_principal, ac.fit_predict(data_principal)))
```

```python
k = list(range(2,7))
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('S(i)', fontsize = 20) 
plt.show() 
```
