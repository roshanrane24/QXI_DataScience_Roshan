```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
import nltk
from textblob import TextBlob
from sklearn.preprocessing import MinMaxScaler
from gensim.summarization import summarize
```

```python
data = pd.read_table('amazon_alexa.tsv')
```

```python
data.head()
```

```python
data2 = pd.get_dummies(data['variation'])
```

```python
data2.head()
```

```python
data = pd.concat([data,data2],axis=1)
```

```python
data.head()
```

```python
data['month'] = pd.DatetimeIndex(data['date']).month
data['day'] = pd.DatetimeIndex(data['date']).day
data = data.drop(['date'],axis=1)
```

```python
data.head()
```

```python
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return TextBlob('hello').sentiment
```

```python
data['sentiment'] = data.verified_reviews.apply(lambda text: sentiment_calc(text))
data.head()
```

```python
data[['polarity', 'subjectivity']] = data['sentiment'].apply(pd.Series)
data = data.drop(['sentiment'], axis=1)
data.head()
```

```python
def avg_word(sentence):
  try:
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))
  except:
    return 0
```

```python
data['number_words'] = data['verified_reviews'].str.split().str.len()
```

```python
data['number_character'] = data['verified_reviews'].str.len()
```

```python
data['avg_word'] = data['verified_reviews'].apply(lambda x: avg_word(x))
```

```python
data.head()
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
data['stopwords'] = data['verified_reviews'].apply(lambda x: len([x for x in x.split() if x in stop]))
data.head()
```

```python
data['stopwords'].describe()
```

```python
scaler = MinMaxScaler()
```

```python
columns = ['number_character','number_words', 'avg_word', 'stopwords']
for col in columns:
  data[[col]] = scaler.fit_transform(data[[col]])
data.head()
```

```python
def sum_text(text):
  try:
    summed_text = summarize(text)
    return summed_text
  except:
    return text
```

```python
data['summed_text'] = data['verified_reviews'].apply(lambda x: sum_text(x))
```

```python
data['sentiment_sum'] = data['summed_text'].apply(lambda text: sentiment_calc(text))
data[['polarity_sum', 'subjectivity_sum']] = data['sentiment_sum'].apply(pd.Series)
data = data.drop(['sentiment_sum'], axis = 1)
data.head()
```

```python
data2 = data.drop(['verified_reviews', 'summed_text'], axis = 1)
data2.head()
```

```python
ns_plot = sns.pairplot(data, hue = 'rating', vars=['polarity','subjectivity', 'number_words', 'number_character'])
```

```python
ns_plot = sns.pairplot(data, hue = 'rating', vars=['polarity','subjectivity', 'polarity_sum','subjectivity_sum'])
```

```python
data2 = data2.drop('variation',axis=1)
```

```python
data_for_model = data2.drop('rating', axis=1)
```

```python
X = data_for_model.iloc[:,1:].values
y = data_for_model.iloc[:,0].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
nbc = GaussianNB()
```

```python
nbc.fit(X_train,y_train)
```

```python
nbc.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
predictions = nbc.predict(X_test)
cm = confusion_matrix(y_test, predictions)
labels = ['1','2','3','4','5']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
print(cm)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
