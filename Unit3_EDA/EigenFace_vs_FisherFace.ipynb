```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from random import randint
import matplotlib.cm as cm
from skimage import io, color
import os
import pandas as pd
from IPython.display import clear_output
from tqdm import tqdm_notebook as tqdm
import pickle
from sklearn.metrics import accuracy_score
```

```python
fpath = 'faces94'
```

```python
# Get total no of files
def totalFileN(fp):
    x = []
    for no in os.listdir(fp):
        for no1 in os.listdir(fp + '/' + no):
                        for no2 in os.listdir(fp + '/' + no + '/' + no1):
                            imgPath  =fp + '/' + no + '/' + no1 + '/' + no2
                            if os.stat(imgPath).st_size > 0:
                                if imgPath.endswith('.jpg') or imgPath.endswith('.gif'):
                                    x.append(no2)
    return len(x),x
```

```python
# Saving Image into array
def subImgArr(path,ImgSize=180*200):
    A = np.empty((1,(ImgSize)))
    y = np.empty([1])
    run ,x= totalFileN(path)
    runt = run
    j = 0
    with tqdm(total=runt) as pbar:
        c = 0
        for no in os.listdir(path):
            if(run <= 0):
                break
            else:
                for no1 in os.listdir(path + '/' + no):
                    j += 1
                    for no2 in os.listdir(path + '/' + no + '/' + no1):
                        imgPath  =path + '/' + no + '/' + no1 + '/' + no2
                        if os.stat(imgPath).st_size > 0:
                            if imgPath.endswith('.jpg') or imgPath.endswith('.gif'):
                                A = np.append(A, color.rgb2gray(io.imread(imgPath)).reshape((1, 180 * 200)),axis=0)
                                y = np.append(y,np.array([j]),axis=0)
                                run -= 1
                                c += 1
                                clear_output()
                                print("{}%".format(round(((runt-run)/runt)*100,2)))
                                print('Working on File: ' + no2)
                                pbar.update(1)
        A = A[1:]
        y = y[1:]
    return A, y
```

```python
# save Image Variables(variables can be list of variables)
def saveVar(var,filename):
    with open(filename,'wb') as f:
        pickle.dump(var,f)
        
# Load Image variables        
def loadVar(var,filename):
    with open(filename,'rb') as f:
        data = pickle.load(f)
        A = data[0]
        y = data[1]
    return A, y
```

```python
A,y = subImgArr(fpath)
```

```python
plt.figure()
pca = PCA(n_components=4)
pca.fit(A)

for i in range(4):
    ax = plt.subplot(2, 2, i + 1)
    ax.imshow(pca.components_[i].reshape([200, 180]))
plt.show()
```

```python
#Colors for distinct individuals
cols = ['#{:06x}'.format(randint(0, 0xffffff)) for i in range(len(np.unique(y)))]
pltCol = [cols[int(k-1)] for k in y]
# %% Plot 2d PCA data
drA = pca.transform(A)
plt.figure()
plt.scatter(drA[:, 0], drA[:, 1], color=pltCol)
plt.show()
```

```python
# %%Compute Fisherfaces
lda = LinearDiscriminantAnalysis()
#Use cross validation to check performance
k_fold = model_selection.KFold(3,shuffle=True)
for (trn, tst) in k_fold.split(range(len(A))):
    #Use PCA to transform from dimension F to dimension N-m
    pca = PCA(n_components=(len(trn) - len(np.unique(y))))
    pca.fit(A[trn])
    #Compute LDA of reduced data
    lda.fit(pca.transform(A[trn]), y[trn])
    yHat = lda.predict(pca.transform(A[tst]))
    #Compute classification error
    outVal = accuracy_score(y[tst], yHat)
    print('Score: ' + str(outVal))
```

```python
# %% Fit all data for plots
pca = PCA(n_components=(len(A) - len(np.unique(y))))
pca.fit(A)
pcatA = pca.transform(A)
lda.fit(pcatA, y)
ldatA = lda.transform(pcatA)
#Plot fisherfaces
plt.figure()
for i in range(4):
    ax = plt.subplot(2, 2, i + 1)
    #Map from PCA space back to the original space (of images)
    C1 = pca.inverse_transform(lda.scalings_[:, i])
    C1.shape = [200, 180]
    ax.imshow(C1)
plt.show()
```
